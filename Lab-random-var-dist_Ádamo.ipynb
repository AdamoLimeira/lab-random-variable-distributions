```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import random
from scipy.stats import bernoulli
from scipy.stats import binom
from scipy.stats import geom
from scipy.stats import poisson
from scipy.stats import uniform
from scipy.stats import expon
from scipy.stats import norm
```

```python
df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
df
```

```python
df.info()
```

```python
numerical = df.select_dtypes(include=[np.number])
numerical
```

```python
plt.figure(figsize=(14, 8))
for i, each in enumerate(numerical, 1):
    plt.subplot(2, 4, i)
    sns.histplot(numerical[each], bins=50, kde=True, color='skyblue')
    plt.xlabel(each, fontsize=12)
    plt.ylabel('Density', fontsize=12)
    plt.title(f'Distribution of {each}', fontsize=10)

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the combined plot
plt.show()
```

```python
def log_transformation(x):
    x = np.log10(x)
    if np.isfinite(x):
        return x
    else:
        return np.NAN
```

```python
# Apply log transformation to each numerical variable
for each in numerical:
    df[each] = df[each].apply(lambda x: log_transformation(x))

# Create a single figure for all distribution plots
plt.figure(figsize=(14, 8))

# Customize and plot distribution plots for each numerical variable
for i, each in enumerate(numerical, 1):
    plt.subplot(2, 4, i)
    sns.histplot(df[each], kde=True, color='orange')
    plt.title(f'Distribution of {each}')
    plt.xlabel(each)
    plt.ylabel('Frequency')

# Adjust layout to prevent overlapping
plt.tight_layout()
```
